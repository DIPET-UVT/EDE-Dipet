```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
data_path = "/Users/Gabriel/Documents/workspaces/diceWorkspace/dmon-adp/data/storm"

df_storm = pd.read_csv(os.path.join(data_path, "AnomalyFrame_Labeled_Complete.csv"))
df_system = pd.read_csv(os.path.join(data_path, "System.csv"))
```

```python
print("Shape of Storm data set is {}".format(df_storm.shape))
df_storm.head()
```

```python
print("Shape of System data set is: {}".format(df_system.shape))
df_system.head()
```

```python
print("Begin Merge of the two dataframes ...")
df_merge = pd.merge(df_storm, df_system, on='key')
print("Shape of merged dataframe is: {}".format(df_merge.shape))
df_merge.head()
```

```python
print("Setting index as key column")
df_merge.set_index('key', inplace=True)
```

```python
len(list(df_merge.columns.values))
```

```python
# print("Saving merged dataframe...")
# df_merge.to_csv(os.path.join(data_path, "All_merged_Storm.csv"))
# print("Done")
```

```python
print("Dropping old target ...")
df_merge.drop(['TargetF'], axis=1, inplace=True)
print("New shape of dataframe is: {}".format(df_merge.shape))
```

```python
df_merge.head()
```

```python
print("Adding anomalies to data")
from sklearn.ensemble import IsolationForest
clf = IsolationForest(max_samples='auto', verbose=1, n_jobs=-1, contamination=0.034)
clf.fit(df_merge)
pred = clf.predict(df_merge)
```

```python
anomalies = np.argwhere(pred == -1)
normal = np.argwhere(pred == 1)
def percentage(part, whole):
  return 100 * float(part)/float(whole)
print("Identified {} anomalous instances, total of {}%.".format(len(anomalies), percentage(len(anomalies), df_merge.shape[0])))
len(anomalies)
len(normal)
```

```python
print("Generating labeled frame ...")
df_merge['TargetF'] = pred
# df_merge.set_index('key', inplace=True)
print("Saving merged dataframe...")
df_merge.to_csv(os.path.join(data_path, "All_merged_Storm.csv"))
print("Done")
# df_merge.to_csv(os.path.join(dataDir, 'Storm_Complete_labeled.csv'))
```

```python
print("Generating anomaly frame ...")
slist = []
for an in anomalies:
    slist.append(df_merge.iloc[an[0]])
df_ano = pd.DataFrame(slist)
# df_ano.set_index('key', inplace=True)
# test.to_csv(os.path.join(dataDir, 'tt.csv'))
print("Saving anomaly dataframe...")
df_ano.to_csv(os.path.join(data_path, "Anomaly_only_Storm.csv"))
print("Done")
```
