```python
def make_overlap_histogram(df, feature, target):
    fig, ax = plt.subplots(figsize=(6, 6))
    g = sns.kdeplot(df[feature][(df[target] == 1) & (df[feature] != -1)], color="Red", shade = True)
    g = sns.kdeplot(df[feature][(df[target] == -1) & (df[feature] != -1)], ax=g, color="Blue", shade= True)
    g.set_xlabel(feature)
    g.set_ylabel("Frequency")
    g = g.legend(["Normal","Anomaly"])
```

```python
continuous_features = ['bolts_0_acked', 'topologyStats_all_transferred',
       'topologyStats_all_emitted', 'topologyStats_3h_emitted',
       'bolts_1_executeLatency', 'bolts_1_processLatency',
       'bolts_1_emitted', 'bolts_1_executed', 'bolts_1_acked',
       'bolts_1_capacity', 'topologyStats_10m_emitted', 'bolts_0_emitted',
       'topologyStats_1d_emitted', 'topologyStats_10m_transferred',
       'topologyStats_3h_transferred', 'topologyStats_1d_transferred',
       'bolts_0_capacity', 'spouts_0_emitted', 'bolts_0_executed',
       'spouts_0_transferred', 'bolts_0_transferred', 'tx_master_x',
       'rx_master_x', 'tx_slave1_x', 'rx_slave1_x', 'tx_slave2_x',
       'rx_slave2_x', 'tx_slave3_x', 'rx_slave3_x', 'shortterm_master',
       'midterm_master', 'longterm_master', 'shortterm_slave1',
       'midterm_slave1', 'longterm_slave1', 'shortterm_slave2',
       'midterm_slave2', 'longterm_slave2', 'shortterm_slave3',
       'midterm_slave3', 'longterm_slave3', 'tx_master_y', 'rx_master_y',
       'tx_slave1_y', 'rx_slave1_y', 'tx_slave2_y', 'rx_slave2_y',
       'tx_slave3_y', 'rx_slave3_y']
```

```python
for feature in continuous_features:
    make_overlap_histogram(df_org, feature, 'TargetF')
```
